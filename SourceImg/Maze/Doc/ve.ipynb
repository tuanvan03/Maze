```python
import pandas as pd
import matplotlib.pyplot as plt

# c d liu t file Excel
df = pd.read_excel("D:\DaiHoc\HK6\TKDGTT\Project\Maze\SourceImg\Maze\Doc\\20_10 to 100_50\\ve.xlsx")
# Lc d liu cho tng thut ton
bfs_data = df[df['Algorithm'] == 'BFS']
dfs_data = df[df['Algorithm'] == 'DFS']
astar_data = df[df['Algorithm'] == 'AStar']
dijkstra_data = df[df['Algorithm'] == 'Dijkstra']
bidirectional_data = df[df['Algorithm'] == 'Bidirectional']

# V biu  ng
plt.figure(figsize=(10, 6))

# V ng cho BFS
plt.plot(bfs_data['Maze'], bfs_data['Runtime (s)'], label='BFS')

# V ng cho DFS
plt.plot(dfs_data['Maze'], dfs_data['Runtime (s)'], label='DFS')

# V ng cho AStar
plt.plot(astar_data['Maze'], astar_data['Runtime (s)'], label='AStar')

# V ng cho Dijkstra
plt.plot(dijkstra_data['Maze'], dijkstra_data['Runtime (s)'], label='Dijkstra')

# V ng cho Bidirectional
plt.plot(bidirectional_data['Maze'], bidirectional_data['Runtime (s)'], label='Bidirectional')

# t tn trc x
plt.xlabel('Vertical Maze')

# t tn trc y
plt.ylabel('Runtime (s)')

# t tiu  cho biu 
plt.title('Comparison of Algorithm Runtimes')

# t ch thch cho biu 
plt.legend()

# plt.xticks(ticks=range(0),rotation=45)


plt.show()
```
