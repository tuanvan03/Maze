```python
import pandas as pd
import matplotlib.pyplot as plt

# c d liu t file Excel
df = pd.read_excel('D:\DaiHoc\HK6\TKDGTT\Project\Maze\SourceImg\Maze\Vuong2\\astar.xlsx')

# Lc d liu cho tng thut ton
bfs_data = df[df['Algorithm'] == 'Astar1']
astar_data = df[df['Algorithm'] == 'AStar']

# V biu  ng
plt.figure(figsize=(10, 6))

# V ng cho BFS
plt.plot(bfs_data['Maze'], bfs_data['Runtime (s)'], label='Astar1')

# V ng cho DFS

# V ng cho AStar
plt.plot(astar_data['Maze'], astar_data['Runtime (s)'], label='AStar')

# V ng cho Dijkstra

# V ng cho Bidirectional

# t tn trc x
plt.xlabel('Square Maze')

# t tn trc y
plt.ylabel('Runtime (s)')

# t tiu  cho biu 
plt.title('Comparison of Algorithm Runtimes')

# t ch thch cho biu 
plt.legend()

# plt.xticks(ticks=range(0),rotation=45)


plt.show()
```
